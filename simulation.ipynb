```python
start_pos = [0.75*math.pi, math.pi, 0.5*math.pi, math.pi, 0.5*math.pi, math.pi]
target_pos = [1.4*math.pi, 1.28*math.pi, 0.67*math.pi, math.pi, 0.63*math.pi, math.pi]
def move_joints(joint_poses, speed = 0.015, precision = 9e-3):
    for i, pos in enumerate(joint_poses):
        if i == 0:
            p.setJointMotorControl2(kinova, 2, p.POSITION_CONTROL, pos, 
                                    targetVelocity = 0, positionGain = speed, 
                                    velocityGain = 1, maxVelocity=4)
        if i == 6:
            p.setJointMotorControl2(kinova, 8, p.POSITION_CONTROL, pos, 
                                    targetVelocity = 0, positionGain = speed, 
                                    velocityGain = 1, maxVelocity=15)
        elif i < 6:
            p.setJointMotorControl2(kinova, i+2, p.POSITION_CONTROL, pos, 
                                    targetVelocity = 0, positionGain = speed, 
                                    velocityGain = 1, maxVelocity=200)
    
    start = time.time()
    while True:
        # Keep moving until joints reach the target configuration
        
        current_joint_state = [p.getJointState(kinova, i+2)[0] for i in range(6)]
        if all([np.abs(current_joint_state[i] - joint_poses[i]) < precision for i in range(6)]): 
            break
        if time.time() - start > 5:
            break
def execute_grasp(position, orientation=p.getQuaternionFromEuler([math.pi,0,math.pi])):
    
    joint_poses = p.calculateInverseKinematics(kinova, 8, position, 
                                               orientation, solver= 0, 
                                               maxNumIterations=150, residualThreshold=1e-4)
    
    hand_open()
    #time.sleep(0.6)
    move_joints(joint_poses)
    #time.sleep(1.5)
    hand_close()
    time.sleep(1.5)
    move_joints(start_pos)
    #time.sleep(1.3)
def set_joint_positions(positions):
    for i in range(6):
        p.resetJointState(kinova, i+2, positions[i])
def collision_check(j, distance=0.15):
        set_joint_positions(j)
        
        hand_close()
        for obstacle_id in obstacles:
            closest_points = p.getClosestPoints(
                kinova, obstacle_id, distance)
            if closest_points is not None and len(closest_points) != 0:
                
                return True
        
        return False
def visualize_path(j_1, j_2, color=[0, 1, 1]):
    # obtain position of first point
    set_joint_positions(j_1)
    point_1 = p.getLinkState(kinova, 8)[0]
    # obtain position of second point
    set_joint_positions(j_2)
    point_2 = p.getLinkState(kinova, 8)[0]
    # draw line between points
    p.addUserDebugLine(point_1, point_2, color, 1.0)


def rrt(j_init, j_goal, iterations, delta_j, steer_to_goal_p, distance=0.15):
    """
   -->input j_init: starting configuration
   -->input j_goal: goal configuration
   -->input iterations: max number of iterations
   -->input delta_j: steer distance
   -->input steer_to_goal_p: probability of steering towards the goal
   -->input distance: distance for collision_check
   <--returns path: list of joint angles
    """
    
    vertices, edges = [j_init], []
    path, found = [], False
    j_rand, j_nearest, j_new = [], [], []
    for i in range(iterations):
        j_rand = semi_random_sample(steer_to_goal_p, j_goal)
        j_nearest = nearest(vertices, j_rand)
        j_new = steer(j_nearest, j_rand, delta_j)
        if not collision_check(j_new, distance):
            if j_new not in vertices:
                vertices.append(j_new)
            if (j_nearest, j_new) not in edges:
                edges.append((j_nearest, j_new))
                visualize_path(j_nearest, j_new)
            if get_euclidean_distance(j_goal, j_new) < delta_j:
                vertices.append(j_goal)
                edges.append((j_new, j_goal))
                visualize_path(j_new, j_goal)
                found = True
                break

    if found:
        
        current_j = j_goal
        path.append(current_j)
        while current_j != j_init:
            for edge in edges:
                if edge[1] == current_j:
                    current_j = edge[0]
                    path.append(edge[0])
        path.reverse()
        return path
    else:
        return None


def semi_random_sample(steer_to_goal_p, j_goal):
    """
   -->input steer_to_goal_p: probability of steering towards the goal
   -->input j_goal: goal configuration

   <--returns j_rand: a uniform random sample of joint angles
    """
    prob = random.random()

    if prob < steer_to_goal_p:
        return j_goal
    else:
        # Uniform random sample over reachable joint angles
        j_rand = [random.uniform(0, 2*np.pi) for i in range(len(j_goal))]
    return j_rand


def get_euclidean_distance(j1, j2):
    distance = 0
    for i in range(len(j1)):
        distance += (j2[i] - j1[i])**2
    return math.sqrt(distance)


def nearest(V, j_rand):
    """
   -->input V: vertices in the current tree
   -->input j_rand: a new uniform random sample

   <--returns j_nearest: the closet point on the tree to j_rand
    """
    # Using euclidean distance
    distance = float("inf")
    j_nearest = None
    for idx, v in enumerate(V):
        if get_euclidean_distance(j_rand, v) < distance:
            j_nearest = v
            distance = get_euclidean_distance(j_rand, v)
    return j_nearest


def steer(j_nearest, j_rand, delta_j):
    j_new = None
    if get_euclidean_distance(j_rand, j_nearest) <= delta_j:
        j_new = j_rand
    else:
        j_hat = [(j_rand[i] - j_nearest[i]) / get_euclidean_distance(j_rand, j_nearest) for i in range(len(j_rand))]
        j_new = [j_nearest[i] + j_hat[i] * delta_j for i in range(len(j_hat))]
    return j_new
```
